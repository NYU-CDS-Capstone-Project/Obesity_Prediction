```python
allIds = []
allTrainTimeSeries = []
allTgts = []

for index in theNewDf.index:
    allIds.append(theNewDf.loc[index, 'ID'])
    tmpTable = np.zeros((embSize, lenTime))
    for i in range(embSize):
        for j in range(lenTime):
            #print(i, j)
            tmpTable[i, j] = theNewDf.loc[index, columnHeaders[i]+'-'+columnSlots[j]]
    allTrainTimeSeries.append(tmpTable)
    allTgts.append(list(theNewTgt2.loc[int(index), :])[1:])
```
