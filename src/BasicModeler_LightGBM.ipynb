```python
dataPath = os.path.abspath('.') + '//dataSourceCsv'
```

```python
columns = {}
columns['predictors'] = [x for x in train.columns if x != 'ID']
columns['ID'] = ['ID']
columns['label'] = ['target_3']
```

```python
np.random.state = 100
```

```python
allIndex = list(range(train.shape[0]))
np.random.shuffle(allIndex)
```

```python
trainIndex = allIndex[:int(len(allIndex)*0.8)]
valIndex = allIndex[int(len(allIndex)*0.8):]
```

```python
train_lgbset = lgb.Dataset(train.loc[trainIndex, columns['predictors']], \
                        label=target.loc[trainIndex, columns['label']], \
                        feature_name=columns['predictors'], \
                        categorical_feature=[],\
                        free_raw_data=False)

val_lgbset = lgb.Dataset(train.loc[valIndex, columns['predictors']], \
                        label=target.loc[valIndex, columns['label']], \
                        feature_name=columns['predictors'], \
                        categorical_feature=[],\
                        free_raw_data=False)
```
